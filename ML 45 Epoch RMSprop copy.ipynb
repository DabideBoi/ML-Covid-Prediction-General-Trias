```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from datetime import timedelta
import datetime as dt

start = dt.date(2020, 3, 20)
end = dt.date(2021, 12, 4)

def daterange(date1, date2):
    for n in range(int ((date2 - date1).days)+1):
        yield date1 + timedelta(n)

def date_init(first, last):
    date_counter = day_counter= week= freq_count= 0
    x, y = [], []
    for days in daterange(first, last):
        for date in xpoints:
            #print(date, " : ", days)
            if date == str(days):
                print(date, " : ", days)
                freq_count = ypoints[date_counter]
                date_counter = date_counter + 1
            
        x.append(days)
        y.append(freq_count)
        freq_count = 0
    return x, y

result = date_init(start, end)
x, y = result[0], result[1]
data = pd.DataFrame( {'Week': x, 'Frequency': y} )
data
```

```python
data.plot(kind="line", x="Week", y="Frequency", figsize=(20, 10))
```

```python
#Machine Learning Starts here
from tensorflow.keras.layers import Dense, Dropout, LSTM
from tensorflow.keras.models import Sequential
from sklearn.preprocessing import MinMaxScaler
```

```python
#Implementing scaler
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data= scaler.fit_transform(data['Frequency'].values.reshape(-1,1))
```

```python
#How many weeks will we base our prediction
predict_weeks = 60
future_week = 30
```

```python
x_train, y_train = [], []

for x in range(predict_weeks, len(scaled_data)-future_week):
    x_train.append(scaled_data[x-predict_weeks:x, 0])
    y_train.append(scaled_data[x + future_week, 0])

x_train, y_train = np.array(x_train), np.array(y_train)
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
```

```python
#Neural Network
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1],1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50))
model.add(Dropout(0.2))
model.add(Dense(units=1))

model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(x_train, y_train, epochs=25, batch_size=32)
```

```python
#Test Model
test_start = dt.date(2020, 3, 20)
test_end = dt.date(2021, 12, 4)
test_result = date_init(test_start, test_end)
test_x, test_y = test_result[0], test_result[1]
test_data = pd.DataFrame( {'Week': test_x, 'Frequency': test_y} )
total_dataset = pd.concat((data['Frequency'], test_data['Frequency']),axis=0)
actual_cases = test_data['Frequency']
```

```python
model_inputs = total_dataset[len(total_dataset) - len(test_data) - predict_weeks:].values
model_inputs = model_inputs.reshape(-1,1)
model_inputs = scaler.fit_transform(model_inputs)
```

```python
test_array = []
for i in range(predict_weeks, len(model_inputs)):
    test_array.append(model_inputs[i-predict_weeks:i, 0])

test_array = np.array(test_array)
test_array = np.reshape(test_array, (test_array.shape[0], test_array.shape[1], 1))
print(test_array)
prediction_cases = model.predict(test_array)
prediction_cases = scaler.inverse_transform(prediction_cases)
```

```python
plt.plot(actual_cases, color='black', label='Actual Cases')
plt.plot(prediction_cases, color='green', label='Predicted Cases')
plt.title("Cases Prediction")
plt.xlabel('Weeks')
plt.ylabel('Cases')

plt.show()
```
